```
import os
import cv2
from google.colab.patches import cv2_imshow

# Caminho da pasta onde as imagens est√£o salvas
folder_path = '/content/yolov5/runs/detect/my_results'

# Listar todos os arquivos na pasta
image_files = [f for f in os.listdir(folder_path) if f.endswith(('.png', '.jpg', '.jpeg'))]

# Exibir todas as imagens
for image_file in image_files:
    print(' ');
    print(image_file);
    # Caminho completo da imagem
    img_path = os.path.join(folder_path, image_file)

    # Ler a imagem
    img = cv2.imread(img_path)

    # Exibir a imagem
    cv2_imshow(img)
    cv2.waitKey(0)  # Espera pela tecla pressionada para passar para a pr√≥xima imagem

cv2.destroyAllWindows()  # Fecha todas as janelas do OpenCV ap√≥s exibir as imagens
```
